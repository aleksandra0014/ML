```python
from sklearn import datasets
import pandas as pd 
import numpy as np
iris = datasets.load_iris(as_frame=True)
```

## IRIS DATASET


```python
data1 = pd.DataFrame(data= np.c_[iris['data'], iris['target']],
                     columns= iris['feature_names'] + ['target'])
```

```python
data1
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
sc.fit(iris['data'])
X = sc.transform(iris['data'])
X
```

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

pca = PCA(n_components=2)
x_reduced = pca.fit_transform(X)
```

```python
plt.scatter(
    x_reduced[:, 0],
    x_reduced[:, 1],
    c=iris.target,
    s=40
)
plt.title('Pierwsze dwie PCA- iris datset')
plt.xlabel('1st PCA')
plt.ylabel('2nd PCA')
```

## LDA

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis

target_names = iris.target_names
y = iris['target']
X = iris['data']
sc3 = StandardScaler()
X = sc3.fit_transform(X)
lda = LinearDiscriminantAnalysis(n_components=2)
X_lda = lda.fit(X, y).transform(X)

plt.figure()
colors = ['navy', 'turquoise', 'darkorange']
lw = 2

for color, i, target_name in zip(colors, [0, 1, 2], target_names):
    plt.scatter(X_lda[y == i, 0], X_lda[y == i, 1], color=color, alpha=.8, lw=lw,
                label=target_name)
plt.legend(loc='best', shadow=False, scatterpoints=1)
plt.title('LDA of IRIS dataset')

plt.show()
```

```python
california = fetch_california_housing()
X = california.data
y = california.target

# PodziaÅ‚ cen domÃ³w na kategorie
low_price = y < 2
medium_price = (y >= 2) & (y < 4)
high_price = y >= 4

# Standaryzacja danych
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Nowe etykiety
new_y = np.zeros_like(y)
new_y[low_price] = 0
new_y[medium_price] = 1
new_y[high_price] = 2

# Redukcja wymiarowoÅ›ci za pomocÄ… LDA
lda = LinearDiscriminantAnalysis(n_components=2)
X_lda = lda.fit(X_scaled, new_y).transform(X_scaled)

# Wykres LDA z kolorowaniem na podstawie kategorii cen
plt.figure()
plt.scatter(X_lda[new_y == 0, 0], X_lda[new_y == 0, 1], c='blue', s=40, label='Niskie ceny')
plt.scatter(X_lda[new_y == 1, 0], X_lda[new_y == 1, 1], c='green', s=40, label='Åšrednie ceny')
plt.scatter(X_lda[new_y == 2, 0], X_lda[new_y == 2, 1], c='red', s=40, label='Wysokie ceny')

# Dodanie legendy
plt.legend()

plt.title('LDA of California Housing dataset')
plt.xlabel('First LDA component')
plt.ylabel('Second LDA component')
plt.show()
```

```python
from sklearn.datasets import load_iris

# ZaÅ‚aduj dane
iris = load_iris()
wine = load_wine()
diabetes = load_diabetes()
california = fetch_california_housing()

# Standaryzacja danych
scaler = StandardScaler()
X_iris = scaler.fit_transform(iris.data)
X_wine = scaler.fit_transform(wine.data)
X_diabetes = scaler.fit_transform(diabetes.data)
X_california = scaler.fit_transform(california.data)

# Obliczenie SVD i redukcja wymiarÃ³w dla kaÅ¼dego zbioru
U_iris, S_iris, Vt_iris = np.linalg.svd(X_iris)
U_wine, S_wine, Vt_wine = np.linalg.svd(X_wine)
U_diabetes, S_diabetes, Vt_diabetes = np.linalg.svd(X_diabetes)
U_california, S_california, Vt_california = np.linalg.svd(X_california)

# Redukcja wymiarÃ³w (wybierz k skÅ‚adowych)
k = 2
X_iris_reduced = np.dot(X_iris, Vt_iris[:k].T)
X_wine_reduced = np.dot(X_wine, Vt_wine[:k].T)
X_diabetes_reduced = np.dot(X_diabetes, Vt_diabetes[:k].T)
X_california_reduced = np.dot(X_california, Vt_california[:k].T)

# Wykresy dla kaÅ¼dego zbioru danych po redukcji wymiarÃ³w
plt.figure(figsize=(12, 10))

plt.subplot(2, 2, 1)
plt.scatter(X_iris_reduced[:, 0], X_iris_reduced[:, 1], c=iris.target)
plt.title('Iris dataset - SVD reduction')
plt.xlabel('Component 1')
plt.ylabel('Component 2')

plt.subplot(2, 2, 2)
plt.scatter(X_wine_reduced[:, 0], X_wine_reduced[:, 1], c=wine.target)
plt.title('Wine dataset - SVD reduction')
plt.xlabel('Component 1')
plt.ylabel('Component 2')

plt.subplot(2, 2, 3)
plt.scatter(X_diabetes_reduced[:, 0], X_diabetes_reduced[:, 1], c=diabetes.target)
plt.title('Diabetes dataset - SVD reduction')
plt.xlabel('Component 1')
plt.ylabel('Component 2')

plt.subplot(2, 2, 4)
plt.scatter(X_california_reduced[:, 0], X_california_reduced[:, 1], c=california.target)
plt.title('California dataset - SVD reduction')
plt.xlabel('Component 1')
plt.ylabel('Component 2')

plt.tight_layout()
plt.show()

```
